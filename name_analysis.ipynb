<h2>What is the distribution of Male and Female names?</h2>


```python
#Initialize the filtered databases for male and female name only
male = pinoyNames[pinoyNames['gender'] == "Male"]
female = pinoyNames[pinoyNames['gender'] == "Female"]

#Plot the bar graph
plt.title("Number of Male and Female Names")
plt.bar(["Male","Female"],[len(male),len(female)])
plt.show()

#Print out the number of male and female names
print("There are over " + str(len(male)) + " Male names and " + str(len(female)) + " Female names.")
```

<h2> What are the the top 25 commonly used names? </h2>


```python
#Initialize the arrays for names and popularity count
nameList, incidence = [], []

#Add the top 25 names to the list via a loop
for x in range(25):
    nameList.append(pinoyNames['forename'].iloc[x])
    value = pinoyNames['incidence'].iloc[x].replace(',', '')  
    incidence.append(int(value))

#Adjust the size of the figure
f = plt.figure()
f.set_figwidth(30)
f.set_figheight(10)

#Plot out the graph and show it
plt.title("Top 25 Most Popular Names")
plt.xlabel("List of Names")
plt.ylabel("Frequency")
plt.bar(nameList, incidence)
plt.show()
```

As you can see from the data above, the top 20 names range from Mary (being the most frequently used name), to Maria, John, Jose, and so forth.</br>
These names are really common althroughout Filipino people. Usually, Filipino names end up being a mix of American and / or Spanish <br>
names, considering our long history with them. <br><br>

Of course, there are different popular names among males and females, and it's best we also have an idea on what they are.

<h2>What are the top 25 common MALE names?</h2>

```python
#Initialize the arrays for the number of male names and their popularity count
maleNameList, maleIncidence = [], []

#Add the top 25 names to the list via a loop
for x in range (25):
    maleNameList.append(male['forename'].iloc[x])
    value = male['incidence'].iloc[x].replace(',', '')  
    maleIncidence.append(int(value))

#Adjust the size of the figure
f = plt.figure()
f.set_figwidth(30)
f.set_figheight(10)

#Plot out the graph and show it
plt.title("Top 25 Most Popular Male Names")
plt.xlabel("List of Names")
plt.ylabel("Frequency")
plt.bar(maleNameList, maleIncidence)
plt.show()
```

No doubt the most common name out there is "John". Trust me when I say, I have met so many Johns throughout my life this far.... too many. <br>
Other name such as Jose, Mark, Michael, and Joel are also widely used. 

<h2>What are the top 25 common FEMALE names?</h2>

```python
#Initialize the arrays for the number of female names and their popularity count
femaleNameList, femaleIncidence = [], []

#Add the top 25 names to the list via a loop
for x in range (25):
    femaleNameList.append(female['forename'].iloc[x])
    value = female['incidence'].iloc[x].replace(',', '')  
    femaleIncidence.append(int(value))

#Adjust the size of the figure
f = plt.figure()
f.set_figwidth(30)
f.set_figheight(10)

#Plot out the graph and show it
plt.title("Top 25 Most Popular Female Names")
plt.xlabel("List of Names")
plt.ylabel("Frequency")
plt.bar(femaleNameList, femaleIncidence)
plt.show()
```

Of course, females also have different popular names. Mary sits at the top with skyrocketing popularity among females.<br>
Other names such as Maria, Jocelyn, Teresita, and others are pretty popular among Filipino children too. <br>
Similar to males, the have a lot of American and Spanish influences. <br><br>

<h2>What are the top 25 least used / unique names?</h2>


If you want to be a bit unique, you might want to consider other names other than "John" and "Mary"...<br><br>

...seriously, stop.<br><br>

There are other names out there that aren't used a lot and actually sound like pretty decent names. As such, these<br>
names could also be a fitting choice for you. If you want some unique names, here are the top 25 unique names you could<br>
consider using.

```python
#Initialize the arrays for the number of unique names and their popularity count
uniqueNameList, uniqueIncidence = [], []

#Add the top 25 names to the list via a loop
for x in range(25):
    uniqueNameList.append(pinoyNames['forename'].iloc[1000-(x+1)])
    value = pinoyNames['incidence'].iloc[1000-(x+1)].replace(',', '')  
    uniqueIncidence.append(int(value))

#Create a function to reverse the list, since the values are initially ordered backwards
def reverseList(list):
    newList = list[::-1]
    return newList

#Adjust the size of the figure
f = plt.figure()
f.set_figwidth(30)
f.set_figheight(10)

#Plot out the graph and show it
plt.title("Top 25 Most Unique Names")
plt.xlabel("List of Names")
plt.ylabel("Frequency")
plt.bar(reverseList(uniqueNameList), reverseList(uniqueIncidence))
plt.show()
```

Among the top 25 unique names, Esteban sits on top while Melecio is the most unique out of all of them.<br>
You also have names like Jovenico, Junie, Ailene, Florita, Teofila, and many others that seem to be mostly<br>
Spanish-influenced, which makes sense. If you want to add some spice to your child's name, these can be worth<br>
a solid shot. 

<h2>Are names more popular if they're longer or shorter?</h2>


Sometimes, certain names bear a certain reputation, and some not so much. Sometimes, you happen to remember<br>
a certain name simply by how short or long it is; it's how some people's minds work. However, does a name's length<br>
actually affect how popular it can get, or at least how well people can remember it?

```python
#Initialize the list for a name's length as well as popularity 
nameLength, incidenceList = [], []

#Add the name length and popularity count for each name in the dataset
for x in range (len(pinoyNames)):
    nameLength.append(len(pinoyNames['forename'].iloc[x]))
    incidenceList.append(pinoyNames['incidence'].iloc[x])

#Adjust the size of the figure
f = plt.figure()
f.set_figwidth(30)
f.set_figheight(10)

#Plot the graph and show it
plt.title("Correlation Plot between Name Length and Popularity")
plt.xlabel("Name Length")
plt.ylabel("Frequency")
plt.scatter(np.array(reverseList(nameLength)), np.array(reverseList(incidenceList)))
plt.show()
```

Surprisingly....no.<br><br>

As you can see, there is not much correlation between names and popularity, so it shouldn't be much of a problem<br>
if you're planning to give a name for your child. 

<h2>How many names are there starting with a specific letter?</h2>


Filipino parents usually have this tradition of having their children's names start with a letter of their own.<br>
It's usually done among siblings, though some parents may also opt to do it directly onto firstborns.<br>
That being said, if you plan to do the same, then you'd best know how many names start with a specific letter.<br>
Take a look at this:

```python
#Generate a complete list of the letters of the alphabet
alphabet = list(string.ascii_uppercase)

#Prepare the array of the number of starting letters 
alphabetLength = []

#For each letter...
#Find out how many names start with a certain latter via a loop
for x in range (len(alphabet)):
    alphabetLength.append(len(pinoyNames[pinoyNames['forename'].str[:1] == alphabet[x]]))

#Adjust the size of the figure
f = plt.figure()
f.set_figwidth(30)
f.set_figheight(10)

#Plot the graph and show it
plt.title("Number of Names per Starting Letter")
plt.xlabel("Starting Letter")
plt.ylabel("Frequency")
plt.bar(alphabet, alphabetLength)
plt.show()
```

Most names start with A or R, while the least amount of people have Y or Z in their name. <br>
Little to no people have Q in their name. Of course, some do, but these are among the most popular ones.<br>
Of course, this shouldn't exactly stop you from naming your child with a starting letter from your name<br>
(if you want to.) This is just for general information.

<h2>What are the top 20 names that start with a specific letter?</h2>

Now that we've covered the number of names starting with a specific letter, what are those names, actually? <br>
If you wish to know, enter the first letter of your first name and find out.

```python
#Prompt and get the user for a letter
letter = input("Enter a letter of your choice: ")

#Filter out the list to names starting with the user's letter
letteredNames = pinoyNames[pinoyNames['forename'].str[:1] == letter.upper()]

#Prepare the list for the list of names as well as their popularity count
letteredNamesList, letteredNamesCount = [], []

#Add the top 20 names into the lists via a loop
for x in range (20):
   letteredNamesList.append(letteredNames['forename'].iloc[x])
   value = letteredNames['incidence'].iloc[x].replace(',', '')  
   letteredNamesCount.append(int(value))
   
#Adjust the size of the figure
f = plt.figure()
f.set_figwidth(30)
f.set_figheight(10)

#Plot out the graph and show it
plt.title("Number of Names per Starting Letter")
plt.xlabel("Starting Letter")
plt.ylabel("Frequency")
plt.bar(letteredNamesList, letteredNamesCount)
plt.show()
```

As you can see, these are the top 20 popular names that begin with the letter of your choice. Pretty neat huh?<br>
Feel free to choose from any of them. Of course, these are only the top 20, so you can have more options aside<br>
from this if you wish.
